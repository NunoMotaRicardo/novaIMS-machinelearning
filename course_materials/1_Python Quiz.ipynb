Functions

```python
x = 10

def modify_x():
    x = 5
    print(x)

print(x)
modify_x()
print(x)

# What will be printed form the cell?
```

```python
def outer_function():
    outer_variable = "Outer"
    def inner_function():
        inner_variable = "Inner"
        print(outer_variable)
        print(inner_variable)
    inner_function()
    print(inner_variable)

outer_function()

# What will this code print?
```

Numpy

```python
import numpy as np

arr1 = np.array([1, 2, 3])
arr2 = arr1
arr2[0] = 9

print(arr1)
print(arr2)

# Ask students: What are the contents of 'arr1' and 'arr2'?
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = arr1 + arr2

print(result)

# What will be the content of 'result'?

```

```python
arr = np.array([1, 2, 3, 4, 5])
filtered_arr = arr[arr > 3]

print(filtered_arr)

# What will be the content of 'filtered_arr'?
```

Pandas

```python
import pandas as pd

df = pd.DataFrame({
    "A": [1, 2, 3],
    "B": ["a", "b", "c"]
})

print(df)



```

```python
# Which columns will the DataFrame 'df' have after the drop operation?
df.drop(columns=["A"])

print(df)
```

```python
df1 = pd.DataFrame({
    "A": [1, 2],
    "B": ["a", "b"]
})

df2 = pd.DataFrame({
    "A": [3, 4],
    "B": ["c", "d"]
})

combined_df = pd.concat([df1, df2])
print(combined_df)

# What will be the content and index values of 'combined_df'?
```
